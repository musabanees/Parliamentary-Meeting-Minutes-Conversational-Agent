Calling API against `User` query

```python
import requests
from IPython.display import display, Markdown
import json

# 1. Configuration
BASE_URL = "http://localhost:8000"
CHAT_ENDPOINT = f"{BASE_URL}/chat"

def ask_agent( query ):
    """
    Sends a query to the Parliamentary Agent API.

    Args:
        query (str): The user's question.

    """
    payload = {
        "query": query
    }

    try:
        response = requests.post(CHAT_ENDPOINT,json=payload)
        response.raise_for_status()  # Raise error for 4xx or 5xx status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}
```

Multi-turn Query Testing

**Scenario 1**

```python
query = """
In the Jan 7, 2025 budget evidence, what did Stephen Boyle say is missing from the forward projections in the budget?
"""
result = ask_agent(query)

if "response" in result:
    display(Markdown(f"Response: {result['response']}"))
    display(Markdown(f"Sources: {set(result.get('sources', []))}"))
```

```python
query = """
Why did he say that matters for responding to shocks and avoiding “playing catch-up”?
"""
result = ask_agent(query)

if "response" in result:
    display(Markdown(f"Response: {result['response']}"))
    display(Markdown(f"Sources: {set(result.get('sources', []))}"))
```
