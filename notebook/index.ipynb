Calling API against `User` query

```python
import requests
from IPython.display import display, Markdown
import json

# 1. Configuration
BASE_URL = "http://localhost:8000"
CHAT_ENDPOINT = f"{BASE_URL}/chat"

def ask_agent( query ):
    """
    Sends a query to the Parliamentary Agent API.

    Args:
        query (str): The user's question.

    """
    payload = {
        "query": query
    }

    try:
        response = requests.post(CHAT_ENDPOINT,json=payload)
        response.raise_for_status()  # Raise error for 4xx or 5xx status codes
        return response.json()
    except requests.exceptions.RequestException as e:
        return {"error": str(e)}
```

```python
query = """
Witnesses across three sessions argue that short-termism undermines sustainability.
"""
result = ask_agent(query)

if "response" in result:
    display(Markdown(f"Response: {result['response']}"))
    display(Markdown(f"Sources: {set(result.get('sources', []))}"))
```
